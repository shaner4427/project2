```python
#Import Dependencies
import pandas as pd
import datetime as dt
```

```python
#Read in CSV
file = pd.read_csv("C:\\Users\\shane\\Downloads\\Police_Incidents.csv")
file.head()
```

```python
#Isolate Columns Needed
file = file[["Division", "Year of Incident", "Call (911) Problem", "Type of Incident", "Time1 of Occurrence", "Day1 of the Week", "Victim Race", "Date of Report", "Modus Operandi (MO)", "Family Offense", "Hate Crime", "Weapon Used", "Gang Related Offense", "Drug Related Istevencident", "Incident Address", "City", "State", "Zip Code", "Location1"]]
file.head()
```

```python
#Isolate Latitude and Longitude from 'Location1' Column
file[["Latitude", "Longitude"]] = (file['Location1']
                                 .str.split('\n')
                                 .apply(pd.Series)[2]  
                                 .str[1:-1]  
                                 .str.split(",", expand=True)  
                                 .astype(float))  
file.head()
```

```python
#Drop Location1 column
file = file.drop(columns=["Location1"])
file.head()
```

```python
#Rename Columns
file = file.rename(columns={
    "Call (911) Problem": "Call Type",
    "Time1 of Occurrence": "Time of Incident",
    "Day1 of the Week": "Day of the Week",
    "Drug Related Istevencident": "Drug Related Incident"
})

file = pd.DataFrame(file)

file.head()
```

```python
#Separate date from 'Date of Report' column
file['Date'] = pd.to_datetime(file['Date of Report']).dt.date

file = file.drop(columns=["Date of Report"])

file.head()
```

```python
#Remove trailing decimal from 'Zip Code' column
file["Zip Code"] = file["Zip Code"].fillna(0).astype(int)

file.head()
```

```python
#Export cleaned data to CSV
file.to_csv("incidents_clean.csv", encoding='utf-8', index=False)
```
